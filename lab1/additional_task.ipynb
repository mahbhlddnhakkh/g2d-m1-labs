```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('lab1/sar_1_gray.jpg') # because I run ipykernel...
image.shape
```

```python
plt.imshow(image)
```

```python
def show_grayscale(image, ax=plt):
    ax.axis('off')
    return ax.imshow(image, cmap="gray", vmin=0, vmax=255)
```

```python
assert((image[:,:,0] == image[:,:,1]).all())
assert((image[:,:,1] == image[:,:,2]).all())
# assert((image[:,:,0] == image[:,:,1] == image[:,:,2]).all()) # doesn't work :c
image = image[:,:,0]
```

```python
show_grayscale(image)
```

```python
def similarity(image1, image2, diff_mean=0.01, diff_std=0.01):
    return int(np.abs(np.mean(image1) - np.mean(image2)) <= diff_mean and np.abs(np.std(image1) - np.std(image2)) <= diff_std)
```

```python
similarity(image, image)
```

```python
similarity(image, image - 0.01)
```

```python
similarity(image, image - 0.1)
```

```python
similarity(image, [0, 1])
```

```python
similarity(image, cv2.imread('lab1/sar_2_color.jpg'))
```
